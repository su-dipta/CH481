# Numerical Differentiation

## $$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def derivative(f,a,method,h):
    if method == 'central':
        return (f(a + h) - f(a - h))/(2*h)
    elif method == 'forward':
        return (f(a + h) - f(a))/h
    elif method == 'backward':
        return (f(a) - f(a - h))/h
    

#f = function f(x)
#a = value at which derivative should be calculated f'(x=a)
#h = the tolerance h, which is a value very close to zero but not zero.     
```

```python
f = np.square
x = np.linspace(0, 4, 41)
num_derivative = derivative(f, x, method='forward',h=0.5)
y = f(x)
print(num_derivative)
true_derivative = 2 * x
plt.figure(figsize=(7,5))
plt.plot(x, y, 'b', label = r'$x^2$')
plt.plot(x, num_derivative, 'r--', label = 'Numerical derivative')
plt.plot(x, true_derivative, 'g', label = 'Analytical derivative')
#num_second = derivative(lambda t: 2*t, x, method='forward', h=0.5)
#plt.plot(x, num_second, 'o', label='second')
plt.grid(True)
plt.legend()
plt.show()
```

```python
derivative(np.cos,0,method='forward',h=0.01)
```

```python
f = np.sin
x = np.linspace(0, 4*np.pi, 100)
num_derivative = derivative(f, x, method='forward',h=0.5)
true_derivative = np.cos(x)
y = f(x)
plt.figure(figsize=(9,6))
plt.plot(x, y, 'b', label='sin(x)')
plt.plot(x, num_derivative, 'o--', label='Numerical derivative')
plt.plot(x, true_derivative, 'g', label='Analytical derivative')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Numerical vs Analytical Derivative of sin(x)")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
f = np.sin

points_list = [10, 20, 100]  # number of points
plt.figure(figsize=(12, 8))

for n in points_list:
    x = np.linspace(0, 4*np.pi, n)
    num_derivative = derivative(f, x, method='forward', h=0.5)
    true_derivative = np.cos(x)
    plt.plot(x, num_derivative, 'o--', label=f'Numerical derivative, n={n}')
    
plt.title("Numerical Derivative of sin(x) with Increasing Points")
plt.xlabel("x")
plt.ylabel("Derivative")
plt.grid('True')
plt.legend()
plt.show()
```

```python
f = lambda t: 0.5 * t ** 3 + 3 * t
t = np.arange(1, 8.5, 0.5)
num_derivative = derivative(f, t, method='forward', h=0.01)
true_derivative = 1.5 * t ** 2 + 3
#num_second = derivative(lambda t:1.5 * t ** 2 + 3,t, method='forward', h=0.01)
y = f(t)
plt.figure(figsize=(9,6))
plt.plot(t, y, 'b', label='f(x)')
plt.plot(t, num_derivative, 'ro--', label='Numerical derivative')
plt.plot(t, true_derivative, 'g', label='Analytical derivative')
#plt.plot(t, num_second, 'o--', label='second')
plt.xlabel('t')
plt.ylabel("f'(t)")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
# define step size
h = 1
# define number of iterations to perform
iterations = 20 
# list to store our step sizes
step_size = [] 
# list to store max error for each step size
max_error = [] 

for i in range(iterations):
    # halve the step size
    h /= 2 
    # store this step size
    step_size.append(h) 
    # compute new grid
    x = np.arange(0, 4 * np.pi, h) 
    # compute function value at grid
    y = np.cos(x) 
    # compute vector of forward differences
    forward_diff = np.diff(y)/h 
    # compute corresponding grid
    x_diff = x[:-1] 
    # compute exact solution
    exact_solution = -np.sin(x_diff) 
    
    # Compute max error between 
    # numerical derivative and exact solution
    max_error.append(\
            max(abs(exact_solution - forward_diff)))

# produce log-log plot of max error versus step size
plt.figure(figsize = (12, 8))
plt.loglog(step_size, max_error, 'v')
plt.show()
```
